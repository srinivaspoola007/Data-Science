### Missing Data

```python
d={'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[4,5,6]}
```

```python
df=pd.DataFrame(d)
df
```

```python
df.dropna() #drop rows containing null values by default
```

```python
df.dropna(axis=1) #drops columns containing null values if we mention axis as 1
```

```python
df.dropna(thresh=2) #drops rows containing atleast 2 null values
```

```python
df.fillna(value='FILL VALUE') #fills the nan value with given value
```

```python
df['A'].fillna(value=df['A'].mean()) #fills the nan value with mean of the column
```

## **merging**

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
left=pd.DataFrame({'key1':['K0','K0','K1','K2'],
                  'key2':['K0','K1','K0','K1'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['K0','K1','K1','K2'],
                    'key2':['k0','k0','k0','k0'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

## **join**

```python
left=pd.DataFrame({'A':['A0','A1','A2'],
                  'B':['B0','B1','B2']},
                 index=['K0','K1','K2'])
right=pd.DataFrame({'C':['C0','C2','C3'],
                  'D':['D0','D2','D3']},
                 index=['K0','K2','K3'])
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```
