```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

# Distribution Plots

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

# **categorical plots**

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
from csv import field_size_limit
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std,width=0.5)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

# Matrix **plots**

```python
flights
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)

g.map_lower(sns.kdeplot)
```

# Regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```
