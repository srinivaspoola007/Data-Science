```python


import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y= x**2
```

```python
x
```

```python
y
```

## **functional**

```python
#functional
plt.plot(x,y)
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Title')
plt.show()
```

```python
#subplotting
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object-Oriented Method

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('Title')
```

```python
fig =plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#subplotting

fig,axes=plt.subplots(nrows=1,ncols=3)
plt.tight_layout()

axes[0].plot(x,y,'b')
axes[0].set_title('first Plot')
axes[1].plot(y,x,'r')
axes[1].set_title('Second Plot')
axes[2].plot(x,y/2,'g')
axes[2].set_title('Third Plot')

```

# **Figure size and DPI**

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,3))
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'g')
plt.tight_layout()
```

```python
#creating a legend

fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x square')
ax.plot(x,x**3,label='x cube')

ax.legend()
```

# Plot Appearance

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='yellow',linewidth=2,linestyle='-',alpha=1,marker='o',markersize=20,markerfacecolor='purple')   #RGB HEX code
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',linewidth=2,linestyle='-.')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
